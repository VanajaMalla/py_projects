```python
df = pd.read_excel('../Py_projects.xlsx')
# df=df.drop('#',axis=1)
# df=df.drop(columns='#')
df
```

```python
# Function to remove content inside parentheses
def remove_parentheses(text):
    return re.sub(r'\(.*?\)', '', text)
```

```python
for  index,row  in df.iterrows() :
    serial_number = row["#"]
    concept=row['Python Concept']
    project =row['Project']

     # Remove content within parentheses from concept and project names
    cleaned_concept = remove_parentheses(concept).strip()
    cleaned_project = remove_parentheses(project).strip()
    
    # Create a directory for the concept with the serial number appended
    concept_dir = f"{serial_number:02d}_{cleaned_concept.replace(' ', '_')}"  # Replace spaces with underscores
    if not os.path.exists(concept_dir):
        os.makedirs(concept_dir)

    # Create a Python file inside the concept folder
    project_filename = cleaned_project.replace(' ', '_') + ".py"  # Replace spaces with underscores for file name
    project_filepath = os.path.join(concept_dir, project_filename)
    # Add Description
    project = project.replace("_"," ").capitalize()
    concept = concept.replace("_"," ").capitalize()
    with open(project_filepath,'w') as file:
        file.write("# Descrption: \n\n# This "+ project + " project is based on python concepts : " + concept)

    print(f"Created file: {project_filepath}")
```
